```python
X = train.drop('price_range',axis=1)
y = train['price_range']
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
train = sc.fit_transform(X)
test = sc.transform(test)
```

```python
train
```

```python
test
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
```

```python
dtc.fit(train , y)
```

```python
pred = dtc.predict(test)
pred
```

```python
dtc.score(train,y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
```

```python
knn.fit(train,y)
```

```python
pred1 = knn.predict(test)
pred1
```

```python
from sklearn.linear_model import LogisticRegression  # its a classification
lr=LogisticRegression()
```

```python
lr.fit(train,y)
```

```python
pred2 = lr.predict(test)
pred2
```

```python
# As we predicted on Test Data csv, we are not able to plot accuracy score as we dont have Ground Truth, so we are going to use only Train.csv and gonna split it into train and test
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train.shape , Y_train.shape , X_test.shape , Y_test.shape
```

```python
dtc1=DecisionTreeClassifier()
```

```python
dtc1.fit(X_train , Y_train)
```

```python
pred4 = dtc1.predict(X_test)
pred4
```

```python
Y_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
dtc1_ac=accuracy_score(pred4,Y_test)
```

```python
dtc1_ac
```

```python
X_train_std=sc.fit_transform(X_train)
X_test_std=sc.transform(X_test)
```

```python
knn1=KNeighborsClassifier()
```

```python
knn1.fit(X_train_std , Y_train)
```

```python
pred5=knn.predict(X_test_std)
pred5
```

```python
knn1_ac=accuracy_score(pred5,Y_test)
```

```python
knn1_ac
```

```python
lr1=LogisticRegression()
```

```python
lr1.fit(X_train_std , Y_train)
```

```python
pred6=lr1.predict(X_test_std)
```

```python
pred6
```

```python
lr1_ac=accuracy_score(pred6,Y_test)
```

```python
lr1_ac
```

```python
plt.bar(x=['dtc1','knn1','lr1'],height=[dtc1_ac,knn1_ac,lr1_ac])
plt.xlabel("Algorithms")
plt.ylabel("Accuracy Score")
plt.show()
```
